```python
#hya new start
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('http://www.nepalstock.com/main/todays_price/index/1/', 'html.parser')
soup=requests.get(url)
# print(soup)
code = soup.text
# print(code)
lxml_code = BeautifulSoup(code, "lxml")
# print(lxml_code)


table_code = lxml_code.table
# tags = table_code.find('tr',class_ = 'unique')

tags = table_code.find_all('tr')
tags



raw_data = []
for tag in tags:
 
    x = tag.text.split('\n')[1:-1]
    if x[0] != "" :
        raw_data.append(x) 
#         print(raw_data)
 


import csv
with open('ram.csv','w') as file:
    x = csv.writer(file)
    x.writerows(raw_data)
    
    
import csv
import pandas as pd
df = pd.read_csv('ram.csv',encoding='latin1')
df    
```
