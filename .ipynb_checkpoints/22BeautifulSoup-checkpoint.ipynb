```python
# Install following
# #beautifulsoup
# request
# lxml
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/', 'html.parser')
soup=requests.get(url)
code = soup.text

lxml_code = BeautifulSoup(code, "lxml")
lxml_code


```

```python
table_code = lxml_code.table
tags = table_code.find_all('tr')
tags
```

```python
raw_data = []
for tag in tags:
    
    x = tag.text.split('\n')[1:-1] #remove un-nacessary data
#     print(x)
    if x[0] != "": #remove empty row
        raw_data.append(x) 
    #raw data save
#     print(tag.text)
# raw_data.split('/n') #list lai split garna mildaina
print(raw_data)
```

```python
import csv
with open('covid.csv','w') as file:
    x = csv.writer(file)
    x.writerows(raw_data)
```

```python
#Data vesulises
import pandas as pd
df = pd.read_csv('covid.csv',encoding='latin1')
df
```

```python
country = list(df['Country,Other'][0:10])
# print(df['Country,Other'][0:10])
total_case = list(df['TotalCases'][0:10])
total_cases = []
for i in total_case:
    x = int(i.replace(',',''))
    total_cases.append(x)
total_cases
```
